## Método de Euler

## Exemplo 1 : y'=y*x com y(0)=1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def odeEuler(f,y0,t):
    
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0,len(t)-1):
        y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
    return y     
```

```python
t = np.linspace(0,1,6)
y0 = 1
f = lambda y,t: y*t
y = odeEuler(f,y0,t)
print(y)
plt.plot(t,y,'b.-')
plt.legend(['Euler'])
plt.axis([0,1,1,2])
plt.grid(True)
plt.show()
```

# dsolve - python

```python
import sympy as sym
import scipy as sp
import matplotlib.pyplot as plt
x = sym.symbols('x')
y= sym.symbols('y', cls=sym.Function)
sol=sym.Eq(y(x).diff(x)-y(x)*x, 0)
sol
```

```python
soln=sym.dsolve(sol,y(x))
soln
```

## Cálculo da constante C1 com condição inicial

```python
constants = sym.solve([soln.rhs.subs(x,0) - 1]) #condição inicial y(0)=1
display(constants)
resposta=soln.subs(constants)
display(resposta)
```

```python
func = sym.lambdify(x,resposta.rhs,'numpy')
xx = sp.arange(0,1,.01) 
y = func(xx)
plt.figure(1)
plt.grid(True)
plt.plot(xx,y);
```

## odeint - python

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# function that returns dy/dt
def model(y,t): 
    dydt = y*t
    return dydt
# initial condition
y0 = 1
# time points
t = np.linspace(0,1,6)
# solve ODE
y = odeint(model,y0,t)
print(y)
# plot results
plt.plot(t,y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.show()
```

## Exemplo 2 : y'=y-(2*x)/y com y(0)=2

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# function that returns dy/dt
def model(y,t): 
    dydt = y-(2*t)/y
    return dydt
# initial condition
y0 = 2
# time points
t = np.linspace(1,1.5,6)
# solve ODE
y = odeint(model,y0,t)
print(y)
# plot results
plt.plot(t,y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.show()
```

## Resolver a equação diferencial y''+5y'+6y=exp(t), com y(0)=1 e y'(0)=3

```python
import sympy as sym
import scipy as sp
import matplotlib.pyplot as plt
t = sym.symbols('t')
y= sym.symbols('y', cls=sym.Function)
sol=sym.Eq(y(t).diff(t,2)+5*y(t).diff(t)+6*y(t), sym.exp(t))
sol
```

```python
soln=sym.dsolve(sol,y(t))
soln
```

```python
sol=soln.rhs
dsol=sol.diff(t)
constants = sym.solve([sol.subs(t,0) - 1, dsol.subs(t,0) - 3] )#condição inicial y(0)=1
display(constants)
resposta=soln.subs(constants)
display(resposta)
```
