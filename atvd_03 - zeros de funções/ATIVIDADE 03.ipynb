```python
#EXERCÍCIO 03

from sympy import *
x=Symbol('x')
def NewtonsMethod(x0):
    tolerance=0.000001
    d=diff(f(x),x)
    while True:
        x1 = x0 - (f(x0)/d.doit().subs({x:x0}))
        t = abs(x1 - x0)
        if t < tolerance:
            break
        x0 = x1
    return x1
def f(x):
    return x**2 + sin(x) - 5
xi = 2.5  # ponto médio do intervalo
xr = NewtonsMethod(xi)
print('ponto médio: ', xi)
print('raiz: ', xr)
print("f(x0) = ", xr**2- sin(xr) - 5)
```
